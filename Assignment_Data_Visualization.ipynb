**Question 5: Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.**

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
# Create a base feature
n_samples = 100
base_feature = np.random.normal(loc=0, scale=1, size=n_samples)

# Create correlated features
feature_1 = base_feature + np.random.normal(loc=0, scale=0.5, size=n_samples)  # Slightly correlated with base_feature
feature_2 = base_feature + np.random.normal(loc=0, scale=0.5, size=n_samples)  # Slightly correlated with base_feature
feature_3 = np.random.normal(loc=0, scale=1, size=n_samples)  # Uncorrelated feature

# Create a DataFrame
data = pd.DataFrame({
    'Base Feature': base_feature,
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar=True)

# Add titles and labels
plt.title("Correlation Matrix Heatmap")
plt.show()
```

**Question 3: Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.**

**np.random.seed(20)**

 **data = {**

 **'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),**

 **'Day': np.random.choice(range(1, 31), 100),**

 **'Sales': np.random.randint(1000, 5000, 100)**

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import plotly.express as px

# Set the random seed for reproducibility
np.random.seed(20)

# Generate the dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table to structure the data for the heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum', fill_value=0)

# Create a heatmap
fig = px.imshow(pivot_table,
                title='Sales Variation Across Months and Days',
                labels=dict(x="Month", y="Day", color="Sales"),
                color_continuous_scale='Viridis')

# Show the plot
fig.show()
```

**Question 4: Using the given x and y data, generate a 3D surface plot to visualize the function.**

**z = sin(sqrt(pow(x,2)+pow(y,2)))**

 **x = np.linspace(-5, 5, 100)**

 **y = np.linspace(-5, 5, 100)**

 **x, y = np.meshgrid(x, y)**

 **z = np.sin(np.sqrt(x**2 + y**2))**

 **data = {**

 **'X': x.flatten(),**

 **'Y': y.flatten(),**

 **'Z': z.flatten()**

 **}**

 **df = pd.DataFrame(data)**

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=df['Z'].values.reshape(100, 100),
                                  x=df['X'].values.reshape(100, 100),
                                  y=df['Y'].values.reshape(100, 100))])

# Update the layout
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(xaxis_title='X',
                             yaxis_title='Y',
                             zaxis_title='Z'),
                  autosize=True)

# Show the plot
fig.show()
```

**Question 2: .Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.**

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values between 0 and 100
y = np.random.rand(n) * 100  # Random y values between 0 and 100
sizes = np.random.randint(10, 100, n)  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh figure
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add scatter renderer to the figure
p.scatter(x='x', y='y', source=data, size='sizes', color='colors', legend_field='colors', alpha=0.6)

# Show the plot
show(p)
```

**Question 4: Create a Bokeh histogram to visualize the distribution of the given data.**

```python
# Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data for the histogram
data_hist = np.random.randn(1000)

# Create a histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Normally Distributed Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           width=800, height=400)

# Add a quad renderer to create the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color='navy', line_color='white')

# Show the plot
show(p)
```

**Question 5:  Create a Bokeh heatmap using the provided dataset.**

**data_heatmap = np.random.rand(10, 10)**

**x = np.linspace(0, 1, 10)**

**y = np.linspace(0, 1, 10)**

**xx, yy = np.meshgrid(x, y)**

```python
# Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import ColorBar, ColumnDataSource

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)

# Create x and y coordinates
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the arrays for plotting
x_flat = xx.flatten()
y_flat = yy.flatten()
data_flat = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data_flat), high=max(data_flat))

# Create a Bokeh figure
p = figure(title="Heatmap",
           x_axis_label='X',
           y_axis_label='Y',
           width=400,
           height=400,
           tools='')

# Add rectangles to the figure
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        fill_color=mapper, line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
